```python
DT=DT.rename({'Delivery Time':'delivery_time', 'Sorting Time':'sorting_time'},axis=1)
```

```python
first_model = smf.ols("delivery_time~sorting_time",data=DT).fit()
```

```python
first_model.summary()
```

```python
model2 = smf.ols('delivery_time~np.log(sorting_time)',data=DT).fit()
model2.summary()
```

```python
model3 = smf.ols('np.log(delivery_time)~sorting_time',data=DT).fit()
model3.summary() 
```

```python
DT["sorting_time_Sq"] = DT.sorting_time**2
model_quad = smf.ols("np.log(delivery_time)~sorting_time+sorting_time_Sq",data=DT).fit()
model_quad.summary()
```

```python
np.exp(model_quad.predict(DT[["sorting_time","sorting_time_Sq"]]))
```
